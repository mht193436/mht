```python
import numpy as np
import matplotlib.pyplot as plt

# 生成数据
x = np.linspace(-5, 5, 1000)  # 在-5到5之间生成1000个均匀分布的点
y = x ** 3  # 计算对应的y值

# 创建图形和坐标轴
plt.figure(figsize=(10, 6))  # 设置图形大小

# 绘制函数曲线
plt.plot(x, y, color='blue', linewidth=2)  # 绘制蓝色曲线，线宽为2

# 添加标题和标签
plt.title('y = x³', fontsize=16)  # 设置标题
plt.xlabel('x', fontsize=14)  # 设置x轴标签
plt.ylabel('y', fontsize=14)  # 设置y轴标签

# 设置坐标轴范围
plt.xlim(-5, 5)  # 设置x轴范围
plt.ylim(-125, 125)  # 设置y轴范围

# 添加网格线
plt.grid(True, linestyle='--', alpha=0.7)  # 添加虚线网格线，透明度为0.7

# 显示图像
plt.tight_layout()  # 自动调整布局
plt.show()  # 显示图形
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

# 设置中文字体
try:
    # 尝试加载系统中的中文字体
    font = FontProperties(fname=r"C:\Windows\Fonts\simhei.ttf")  # Windows系统默认黑体
except:
    # 如果找不到指定字体，使用matplotlib支持的中文字体
    plt.rcParams["font.family"] = ["SimHei", "WenQuanYi Micro Hei", "Heiti TC"]
    font = None

# 生成日期数据
years = np.arange(2000, 2021)
months = np.random.randint(1, 13, size=20)
days = np.random.randint(1, 31, size=20)
dates = [f"{y}/{m:02d}/{d:02d}" for y, m, d in zip(years, months, days)]

# 生成随机销量数据
sales = np.random.randint(500, 2000, size=len(dates))

# 创建图形
plt.figure(figsize=(12, 6))  # 设置图形大小
plt.plot(dates, sales, 'o-', color='blue', linewidth=2, markersize=6)  # 绘制带标记的折线图

# 设置标题和坐标轴标签
plt.title("销售数据趋势图", fontproperties=font, fontsize=16)
plt.xlabel("日期", fontproperties=font, fontsize=14)
plt.ylabel("销量", fontproperties=font, fontsize=14)

# 设置x轴刻度和标签旋转
plt.xticks(range(0, len(dates), 2), [f'%s' % d for d in dates[::2]], rotation=45, fontproperties=font)

# 添加网格线
plt.grid(True, linestyle='--', alpha=0.7)

# 调整布局
plt.tight_layout()

# 显示图形
plt.show()
```
